```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# Load the image (replace 'img_path' with the path to your image)
img = cv2.imread('data/train/infected/img_0_993.jpg', cv2.IMREAD_GRAYSCALE)

# Apply binary thresholding
ret, thresh1 = cv2.threshold(img,10,50, cv2.THRESH_BINARY)

# Create a custom colormap to highlight thresholded values in red
cmap = mcolors.ListedColormap(['red','grey'])

# Normalize thresholded image to [0, 1] for the custom colormap
#normed_thresh1 = (thresh1 - thresh1.min()) / (thresh1.max() - thresh1.min())

# Display the thresholded image with the custom colormap
plt.imshow(thresh1, cmap=cmap)
plt.title('Thresholded Image (Thresholded values in red)')
plt.colorbar()  # Add a colorbar to show the mapping
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the image (replace 'img_path' with the path to your image)
img = cv2.imread('data/train/infected/img_0_993.jpg', cv2.IMREAD_GRAYSCALE)

# Apply binary thresholding to create a mask for black regions
ret, mask = cv2.threshold(img, 10,30, cv2.THRESH_BINARY)

# Create a copy of the original image
highlighted_img = img.copy()

# Set the non-black regions in the copy to a different color (e.g., white)
highlighted_img[mask == 0] = 255  # 255 is white in grayscale

# Display the image with black regions and non-black regions highlighted
plt.imshow(highlighted_img, cmap='gray')
plt.title('Original Image with Black Regions Highlighted')
plt.show()
```
